```python
#there are 5 subjects psychology,nano technology,medicine,oncology,anthropology.2 students have applied to do masters in any of these 5 fields.with a line plot show in which field they can excel.The cutoff for each field is 70.
#[67,45,73,71,57]-->student 1
#[43,67,68,53,85-->student2]
#plotting
sub =["psychology","nanotechnology","medicine","oncology","anthropology"]
S1=[67,45,73,71,57]
cut=[70,70,70,70,70]
S2=[43,67,68,53,85]
#student1
plt.plot(sub,S1,label="Student1")
#student2
plt.plot(sub,S2,label="Student2")
#cutoff
plt.plot(sub,cut,label="Cutoff")
plt.legend()
```

```python
#India vs Australia match was on live.India has given 298 as target for Australia.The runs for every 5 overs is given as follows
#India--[25,51,84,131,160,189,220,250,267,297]
#Australia has completed only for 25 overs(drinks break)
#Australia--[15,41,94,110,151]
#plot overs vs runs for both the countries with India as blue  line and Australia as yelllow line.Give appropriate labels & legends.The marker for India is flag & for Australia is +.Take overs from arange() function.Create runs for arrays
import numpy as np
ind=np.array([25,51,84,131,160,189,220,250,267,297])
over1=np.arange(5,51,5)
aus=np.array([15,41,94,110,151])
over2=np.arange(5,26,5)
#plt.plottype(X,Y,label,color,marker,line width,line style)
plt.plot(over1,ind,label="India",color="b",marker=">",linewidth=3)
plt.plot(over2,aus,label="Australia",color="y",marker="P",linewidth=3)
plt.title("IND vs AUS")
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.grid()
plt.legend()
```
