```python
#hypothesis testing:
#importing required modules
import numpy as np
from scipy import stats
```

```python
#2 groups of data
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
#Perform t_test assuming unequal vaiances
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#print the results
print("T-statistic:",t_stat)
print("P-value:",p_value)
```

```python
#assign alpha value
alpha=0.05
```

```python
#Interpret the results
if p_value< alpha:
  print("Reject the null hypothesis.The means are significant")
else:
  print("Failed to reject the null hypothesis.The means are not significant")
```

REGRESSION ANALYSIS!!!!

```python
import pandas as pd
df=pd.read_csv("/content/Vaccine.csv",sep=",")
```

```python
df
```

```python
a=df["Efficiency"].mean()
print("mean efficiency",a)
b=df["Rec_Rate"].mean()
print("mean rc_rate",b)
data1=df["Efficiency"]
data2=df['Rec_Rate']
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
print(t_stat)
print(p_value)
```

```python
alpha=0.05
```

```python
#Interpret the results
if p_value< alpha:
  print("H0-realated or not different")
else:
  print("h1- not related or different")
```

ANOVA TESTING(testing based on f value)

(One-wayANOVA)

```python
dfA=df[df['Vaccine']=='A']['Efficiency']
dfB=df[df['Vaccine']=='B']['Efficiency']
dfA
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)
print("F-statistics:",f_stat)
print("p_value:",p_value)
```

```python
alpha=0.05
```

```python
#Interpret the results
if f_stat< alpha:
  print("H0-realated or not different")
else:
  print("h1- not related or different")
```
