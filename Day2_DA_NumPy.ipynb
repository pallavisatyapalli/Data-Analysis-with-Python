```python
#joining arrays
A=np.array([34,56,78,90]).reshape(2,2)
B=np.array([3,7,9,12]).reshape(2,2)
print(A)
print("---------------------")
print(B)  #axis=0 --->columns
print("After vertically stacking the arrays!!!")
print(np.vstack((A,B)))  #axis=1 -->row

```

```python
#joining arrays using dstack
A=np.arange(30).reshape(2,3,5)  #no.of rows==no.of groups
print(A)                        #no.of columns in every group==no.of rows in every group
                                       # groups  rows  columns
print("After dstack")           #original 2      3      5
print(np.dstack(A))             #dstack   3      5      2
```

JOINS

```python
A=np.array([1,2,3])
B=np.array([4,5,6])
print(np.inner(A,B))  # the output is given like sum of product of elements-->[(1*4)+(2*5)+(3*6)]=32
```

```python
A=np.array([1,2,3])
B=np.array([4,5,6])
print(np.outer(A,B))  # the output is given like sum of product of elements-->[(1*4)+(1*5)+(1*6)]->row1
                                                                            #  [(2*4)+(2*5)+(2*6)]->row2
                                                                            #[(3*4)+(3*5)+(3*6)]->row3
```

```python
A=np.array([1,2,3])
B=np.array([4,5,6])
print(np.cross(A,B))  # the output is determinant
```
