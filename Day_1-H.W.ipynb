```python
#mytuple = ("Elements")
#creating elements in a tuple
t = ("apple", "banana", "cherry")
print(t)
```

```python
#Tuples allow duplicate values
t = ("apple", "banana", "cherry", "apple", "cherry")
print(t)
```

```python
#finding length of tuple
t = ("apple", "banana", "cherry")
print(len(t))
```

```python
#One item tuple,
t = ("apple",)
print(type(t))

#NOT a tuple
t = ("apple")
print(type(t))
```

```python
#accessing tuples
t= ("apple", "banana", "cherry")
print(t[1])
```

```python
#accessing tuples with negative index
t = ("apple", "banana", "cherry")
print(t[-1])
```

```python
t = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(t[2:5])
```

```python
#removing elements
t= ("apple", "banana", "cherry")
y = list(t)
y.remove("apple")
t= tuple(y)
print(t)
```

```python
t= ("apple", "banana", "cherry")
del t
print(t)
```

```python
t= ("apple", "banana", "cherry")
y = list(t)
y.append("orange")
t= tuple(y)
print(t)
```

```python
#adding elements
s= {"apple", "banana", "cherry"}
s.add("orange")
print(s)
```

```python
#updating elements
s = {"apple", "banana", "cherry"}
s1= {"pineapple", "mango", "papaya"}
s.update(s1)
print(s)
```

```python
#removing elements
s = {"apple", "banana", "cherry"}
s.remove("banana")
print(s)
```

```python
#discarding elements
s = {"apple", "banana", "cherry"}
s.discard("banana")
print(s)
```
