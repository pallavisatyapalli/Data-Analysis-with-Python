EXTRACTING A CONTENT FROM WEBPAGE---RICHEST MEN IN INDIA FORBES

```python
u=("https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1")
```

```python
p=requests.get(u)
```

```python
c=BeautifulSoup(p.text,'html')
print(c)
```

```python
t1=c.find('table')
print(t1)
```

```python
h=t1.find_all('th')
print(h)
```

```python
Headings=[i.text for i in h]
print(Headings)
```

H->consists of text

h->consists of tags

```python
import pandas as pd
```

```python
newdf=pd.DataFrame(columns=Headings)
print(newdf)
```

```python
rows=t1.find_all('tr')
print(rows)
```

```python
data=t1.find_all('td')
print(data)
```

```python
D=[i.text for i in data]
print(D)
```

```python
l=len(D)
print(l)
```

```python
mydf1=pd.DataFrame(columns=D)
print(mydf1)
```

```python
D[0:4]
```

```python
st=0
sp=4
ind=0
for i in D:
  if sp<40:
    newdf.loc[ind]=D[st:sp]
    st=st+4
    sp=sp+4
    ind=ind+1
```
